```python
import pandas as pd
import seaborn as sns

# --- Part 1: Load a Dataset & Check Missing Values ---

print("--- Part 1: Loading Dataset & Checking Missing Values ---")

# Task 1: Titanic Dataset - Load the dataset
# You can load it from a CSV file if you have it locally:
# df = pd.read_csv('titanic.csv')
# For this example, we'll use the dataset provided by Seaborn
try:
    df = sns.load_dataset('titanic')
    print("Titanic dataset loaded successfully.")
except Exception as e:
    print(f"Error loading Titanic dataset: {e}")
    print("Please ensure you have an internet connection or the 'titanic.csv' file locally.")
    exit() # Exit if dataset cannot be loaded

# Check for missing values in the 'Age' column
missing_age_values = df['Age'].isnull().sum()
print(f"\nNumber of missing values in the 'Age' column: {missing_age_values}")

# --- Part 2: Identify Duplicates & Inconsistencies ---

print("\n--- Part 2: Identifying Duplicate Rows ---")

# Task 2: Duplicate Rows in Titanic Dataset
# Identify any duplicate rows in the dataset
duplicate_rows = df.duplicated().sum()
print(f"Number of duplicate rows in the dataset: {duplicate_rows}")

# --- Part 3: Generate a Data Quality Report ---

print("\n--- Part 3: Generating Data Quality Report ---")

# Task 3: Titanic Dataset Overview
print("\n--- Data Quality Report for Titanic Dataset ---")

print("\n1. Missing Values:")
# Calculate missing values for all columns
missing_values_report = df.isnull().sum()
# Filter out columns with no missing values for a cleaner report
missing_values_report = missing_values_report[missing_values_report > 0].sort_values(ascending=False)

if not missing_values_report.empty:
    print(missing_values_report)
else:
    print("No missing values found in any column.")

print(f"\n2. Duplicate Rows: {duplicate_rows}")

print("\n3. Basic Statistics (Numerical Columns):")
print(df.describe())

print("\n4. Basic Statistics (Categorical/Object Columns):")
# Include object type columns for their basic statistics
print(df.describe(include=['object', 'category']))

print("\n--- End of Data Quality Report ---")
```

```python
import pandas as pd

# Load Titanic dataset from seaborn (or replace with your CSV path)
import seaborn as sns
titanic = sns.load_dataset('titanic')

# Check for missing values in the 'age' column
missing_age_count = titanic['age'].isna().sum()

print(f"Number of missing values in 'age' column: {missing_age_count}")
```

```python
# Check for duplicate rows
duplicates = titanic.duplicated()

# Count number of duplicates
num_duplicates = duplicates.sum()

print(f"Number of duplicate rows: {num_duplicates}")

# Optionally, display duplicate rows if any
if num_duplicates > 0:
    print("Duplicate rows:")
    print(titanic[duplicates])
```

```python
# Missing values count per column
missing_values = titanic.isna().sum()

# Number of duplicate rows
duplicates_count = titanic.duplicated().sum()

# Basic statistics summary
summary_stats = titanic.describe(include='all')

print("Missing Values per Column:")
print(missing_values)

print("\nNumber of Duplicate Rows:", duplicates_count)

print("\nSummary Statistics:")
print(summary_stats)
```
