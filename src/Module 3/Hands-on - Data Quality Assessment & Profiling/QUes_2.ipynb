```python
import pandas as pd

# Load Titanic dataset from seaborn (or replace with your CSV path)
import seaborn as sns
titanic = sns.load_dataset('titanic')

# Check for missing values in the 'age' column
missing_age_count = titanic['age'].isna().sum()

print(f"Number of missing values in 'age' column: {missing_age_count}")
```

```python
# Check for duplicate rows
duplicates = titanic.duplicated()

# Count number of duplicates
num_duplicates = duplicates.sum()

print(f"Number of duplicate rows: {num_duplicates}")

# Optionally, display duplicate rows if any
if num_duplicates > 0:
    print("Duplicate rows:")
    print(titanic[duplicates])
```

```python
# Missing values count per column
missing_values = titanic.isna().sum()

# Number of duplicate rows
duplicates_count = titanic.duplicated().sum()

# Basic statistics summary
summary_stats = titanic.describe(include='all')

print("Missing Values per Column:")
print(missing_values)

print("\nNumber of Duplicate Rows:", duplicates_count)

print("\nSummary Statistics:")
print(summary_stats)
```
