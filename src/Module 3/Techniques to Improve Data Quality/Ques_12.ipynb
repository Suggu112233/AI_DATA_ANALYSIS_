```python
# Activity 2: Dealing with Duplicates & Redundancy

# Task A: Identifying Duplicate Records

# 7. Identify Complete Duplicates:
# - Load a dataset and identify duplicated rows.
# - Use Pandas to detect duplicates.






# 8. Identify Duplicates based on Specific Columns:
# - Check for duplicates in specified columns.







# 9. Count Duplicate Rows:
# - Calculate and print the number of duplicate rows.





```

```python
import pandas as pd

# Sample dataset with duplicates
data = {
    'ID': [1, 2, 3, 4, 2, 5, 3],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Bob', 'Eve', 'Charlie'],
    'Email': ['alice@example.com', 'bob@example.com', 'charlie@example.com',
              'david@example.com', 'bob@example.com', 'eve@example.com', 'charlie@example.com']
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# --- 7. Identify Complete Duplicate Rows ---
duplicates = df[df.duplicated()]
print("\nComplete Duplicate Rows:")
print(duplicates)

# --- 8. Identify Duplicates based on Specific Columns (e.g., 'Name' and 'Email') ---
duplicates_specific = df[df.duplicated(subset=['Name', 'Email'])]
print("\nDuplicate Rows based on 'Name' and 'Email':")
print(duplicates_specific)

# --- 9. Count Total Number of Duplicate Rows ---
num_duplicates = df.duplicated().sum()
print(f"\nNumber of complete duplicate rows: {num_duplicates}")
```

```python
# Task B: Deduplication Techniques

# 10. Remove Complete Duplicates:
# - Drop duplicate rows and keep only the first occurrence.






# 11. Subset Deduplication:
# - Remove duplicates based on a subset of columns.






# 12. Keep Last Occurrence:
# - Drop duplicates but keep the last occurrence in the dataset.





```

```python
import pandas as pd

# Sample dataset with duplicates
data = {
    'ID': [1, 2, 3, 4, 2, 5, 3],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Bob', 'Eve', 'Charlie'],
    'Email': ['alice@example.com', 'bob@example.com', 'charlie@example.com',
              'david@example.com', 'bob@example.com', 'eve@example.com', 'charlie@example.com']
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# --- 7. Identify Complete Duplicate Rows ---
duplicates = df[df.duplicated()]
print("\nComplete Duplicate Rows:")
print(duplicates)

# --- 8. Identify Duplicates based on Specific Columns (e.g., 'Name' and 'Email') ---
duplicates_specific = df[df.duplicated(subset=['Name', 'Email'])]
print("\nDuplicate Rows based on 'Name' and 'Email':")
print(duplicates_specific)

# --- 9. Count Total Number of Duplicate Rows ---
num_duplicates = df.duplicated().sum()
print(f"\nNumber of complete duplicate rows: {num_duplicates}")

# --- Additional: Remove complete duplicates ---
df_no_duplicates = df.drop_duplicates()
print("\nDataFrame after removing complete duplicates:")
print(df_no_duplicates)

# --- Additional: Remove duplicates based on 'Name' and 'Email' ---
df_no_duplicates_specific = df.drop_duplicates(subset=['Name', 'Email'])
print("\nDataFrame after removing duplicates based on 'Name' and 'Email':")
print(df_no_duplicates_specific)
```
