```python
# Styling with Matplotlib - Line Plot
# Objective: Create a line plot with customized styles using Matplotlib.
```

```python
import matplotlib.pyplot as plt

# Sample data: months vs sales
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [150, 180, 170, 220, 210, 250]

# Create the line plot with custom styles
plt.plot(months, sales, color='purple', linestyle='--', marker='o', markersize=8, linewidth=2, markerfacecolor='orange')

# Add title and labels with custom fonts
plt.title('Monthly Sales Trend', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales (Units)', fontsize=12)

# Add grid with custom style
plt.grid(True, linestyle='-.', linewidth=0.7, alpha=0.7)

# Show the plot
plt.show()
```
