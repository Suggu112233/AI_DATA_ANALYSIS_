```python
# Combining Plots with Matplotlib
# Objective: Combine multiple plots into a single figure using Matplotlib's subplots.
```

```python
import matplotlib.pyplot as plt

# Sample data
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
monthly_sales = [150, 180, 170, 220, 210, 250]

# Create subplots: 1 row, 2 columns
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Bar plot on the first subplot
axs[0].bar(products, sales, color='skyblue')
axs[0].set_title('Product Sales')
axs[0].set_xlabel('Product')
axs[0].set_ylabel('Sales (in thousands)')
axs[0].grid(True, linestyle='--', alpha=0.7)

# Line plot on the second subplot
axs[1].plot(months, monthly_sales, marker='o', color='orange', linestyle='-')
axs[1].set_title('Monthly Sales Trend')
axs[1].set_xlabel('Month')
axs[1].set_ylabel('Sales (Units)')
axs[1].grid(True, linestyle='--', alpha=0.7)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the combined plot
plt.show()
```
