```python
# Matplotlib Subplots for Multiple Data Sets
# Description: Create subplots using Matplotlib to display the bar plot for product sales and the scatter plot for
# height vs. weight side by side.
```

```python
import matplotlib.pyplot as plt

# Data for bar plot (product sales)
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

# Data for scatter plot (height vs. weight)
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]

# Create subplots: 1 row, 2 columns
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Bar plot
ax1.bar(products, sales, color='skyblue')
ax1.set_title('Product Sales (in Thousands)', fontsize=12, fontweight='bold')
ax1.set_xlabel('Products')
ax1.set_ylabel('Sales')
ax1.grid(True, linestyle='--', alpha=0.6, axis='y')

# Scatter plot
ax2.scatter(heights, weights, color='green')
ax2.set_title('Height vs Weight', fontsize=12, fontweight='bold')
ax2.set_xlabel('Height (cm)')
ax2.set_ylabel('Weight (kg)')
ax2.grid(True, linestyle='--', alpha=0.6)

# Adjust layout for better spacing
plt.tight_layout()

# Display the plots
plt.show()
```
