```python
# Customizing Plots with Seaborn - Boxplot
# Objective: Customize a boxplot with Seaborn to show the distribution of petal lengths by
# species.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load iris dataset
iris = sns.load_dataset('iris')

# Create a boxplot of petal length by species with customizations
sns.boxplot(x='species', y='petal_length', data=iris, palette='Set3', linewidth=2, fliersize=6)

# Add title and labels with custom font sizes and weight
plt.title('Distribution of Petal Length by Species', fontsize=16, fontweight='bold')
plt.xlabel('Species', fontsize=12)
plt.ylabel('Petal Length (cm)', fontsize=12)

# Optionally add grid for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
