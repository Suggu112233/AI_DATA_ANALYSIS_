```python
# Seaborn Plot - Heatmap
# Objective: Create a heatmap to visualize a correlation matrix with Seaborn, using
# advanced styling.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset('iris')

# Calculate the correlation matrix
corr = iris.corr()

# Create the heatmap with advanced styling
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0,
            linewidths=0.5, linecolor='gray', cbar_kws={'shrink': 0.8, 'label': 'Correlation Coefficient'})

# Add title
plt.title('Correlation Matrix Heatmap - Iris Dataset', fontsize=16, fontweight='bold')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load iris dataset
iris = sns.load_dataset('iris')

# Calculate correlation matrix only for numeric columns
corr = iris.select_dtypes(include='number').corr()

# Create the heatmap
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0,
            linewidths=0.5, linecolor='gray', cbar_kws={'shrink': 0.8, 'label': 'Correlation Coefficient'})

plt.title('Correlation Matrix Heatmap - Iris Dataset', fontsize=16, fontweight='bold')
plt.show()
```
