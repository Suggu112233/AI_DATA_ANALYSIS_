```python
# Title: Basic Data Visualization
# Description: Plot the distribution of one feature using Seaborn.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset('iris')

# Plot the distribution of sepal length
sns.histplot(iris['sepal_length'], kde=True, color='skyblue')

# Add title and labels
plt.title('Distribution of Sepal Length', fontsize=16, fontweight='bold')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
