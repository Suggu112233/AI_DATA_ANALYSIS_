```python
# Step 1: Set Up the Environment
# Objective: Ensure you have the right tools and packages installed.

# 1. Install the required libraries.
# 2. Verify the installation by importing the libraries in a Python script or Jupyter notebook





# Step 2: Load & Explore the Dataset
# Objective: Load data into a pandas DataFrame and obtain a basic understanding of its structure.

# 3. Load a CSV file into a DataFrame.
# 4. Display the first few records to understand the structure.
# 5. Get a summary of the dataset.






# Step 3: Perform NumPy Operations
# Objective: Utilize NumPy for basic numerical operations and array manipulations.

# 6. Convert a DataFrame column to a NumPy array and perform array operations like mean and sum.
# 7. Create a NumPy array and calculate the variance and standard deviation.
# 8. Use NumPy to filter based on conditions.








# Step 4: Data Manipulation with Pandas
# Objective: Use Pandas to clean and manipulate dataset for analysis.

# 9. Handle missing data by filling or dropping.
# 10. Create new columns or modify existing ones.
# 11. Use groupby to aggregate data.







# Step 5: Data Visualization with Matplotlib & Seaborn
# Objective: Visualize the data to identify patterns, trends, and insights.

# 12. Use Matplotlib to create a basic plot.
# 13. Create a histogram using Seaborn.
# 14. Plot a box plot for a clear view of data distribution.


# Step 1: Set Up the Environment
# Objective: Ensure you have the right tools and packages installed.

# 1. (Optional) Install required libraries
# !pip install pandas numpy matplotlib seaborn

# 2. Import libraries
# -----------------------------------------------------------------------------
# Full Script: Manually Create CSV + Load, Analyze & Visualize
# -----------------------------------------------------------------------------

# Step 0: Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# -----------------------------------------------------------------------------
# Part A: Create the CSV file (run this once at the top)
# -----------------------------------------------------------------------------

# 1. Define sample data for grouped_mean_by_species.csv
data = {
    'species': ['Setosa', 'Versicolor', 'Virginica'],
    'mean_weight': [45.6, 52.3, 60.8]
}

# 2. Build DataFrame and save to CSV
df_gen = pd.DataFrame(data)
csv_filename = 'grouped_mean_by_species.csv'
df_gen.to_csv(csv_filename, index=False)
print(f"âœ” Created '{csv_filename}' with sample data:\n", df_gen, "\n")

# -----------------------------------------------------------------------------
# Part B: Load & Explore the Dataset
# -----------------------------------------------------------------------------

# 3. Confirm working directory and file exists
print("Current working directory:", os.getcwd())
print("File exists? ", os.path.exists(csv_filename), "\n")

# 4. Load CSV into DataFrame
df = pd.read_csv(csv_filename)

# 5. Display the first few records
print("First few rows of the dataset:")
print(df.head(), "\n")

# 6. Get a concise summary of the DataFrame
print("Summary of the dataset:")
df.info()
print("\n")

# -----------------------------------------------------------------------------
# Part C: Perform NumPy Operations
# -----------------------------------------------------------------------------

# 7. Convert 'mean_weight' column to NumPy array and compute mean & sum
weights = df['mean_weight'].to_numpy()
print("NumPy array (mean_weight):", weights)
print("Mean of mean_weight:", np.mean(weights))
print("Sum of mean_weight:", np.sum(weights), "\n")

# 8. Manual NumPy array variance & std deviation
manual = np.array([10, 20, 30, 40, 50])
print("Manual array:", manual)
print("Variance:", np.var(manual))
print("Standard Deviation:", np.std(manual), "\n")

# 9. Filter rows where mean_weight > 50
filtered = df[df['mean_weight'] > 50]
print("Rows with mean_weight > 50:")
print(filtered, "\n")

# -----------------------------------------------------------------------------
# Part D: Data Manipulation with Pandas
# -----------------------------------------------------------------------------

# 10. Check for and fill missing values in 'mean_weight'
print("Missing data per column before fill:")
print(df.isnull().sum(), "\n")
df['mean_weight'].fillna(df['mean_weight'].mean(), inplace=True)

# 11. Create a new column 'scaled_weight'
df['scaled_weight'] = df['mean_weight'] / df['mean_weight'].max()
print("DataFrame with new 'scaled_weight' column:")
print(df, "\n")

# 12. Group by 'species' and compute mean of 'mean_weight'
grouped = df.groupby('species')['mean_weight'].mean().reset_index()
print("Grouped mean weight by species:")
print(grouped, "\n")

# -----------------------------------------------------------------------------
# Part E: Data Visualization with Matplotlib & Seaborn
# -----------------------------------------------------------------------------

# 13. Basic bar plot: species vs. mean_weight
plt.figure(figsize=(8, 5))
plt.bar(df['species'], df['mean_weight'], color='skyblue')
plt.xlabel("Species", fontsize=12, fontweight='bold')
plt.ylabel("Mean Weight", fontsize=12, fontweight='bold')
plt.title("Mean Weight by Species", fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.grid(True, axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# 14. Histogram of mean_weight using Seaborn
plt.figure(figsize=(6, 4))
sns.histplot(df['mean_weight'], bins=5, kde=True)
plt.title("Distribution of Mean Weight")
plt.xlabel("Mean Weight")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# 15. Box plot of mean_weight
plt.figure(figsize=(6, 4))
sns.boxplot(y=df['mean_weight'])
plt.title("Box Plot of Mean Weight")
plt.ylabel("Mean Weight")
plt.tight_layout()
plt.show()


```

```python
df = sns.load_dataset('iris')
```

```python
arr = df['sepal_length'].to_numpy()  # Example column
print("Mean:", np.mean(arr))
print("Sum:", np.sum(arr))
```

```python
arr2 = df['sepal_width'].to_numpy()
print("Variance:", np.var(arr2))
print("Standard Deviation:", np.std(arr2))
```

```python
filtered = arr[arr > 5.0]
print("Values greater than 5.0:", filtered)
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
```

```python
df['petal_length_to_width_ratio'] = df['petal_length'] / df['petal_width']
```

```python
grouped = df.groupby('species').mean()
print(grouped)
```

```python
grouped['sepal_length'].plot(kind='bar')
plt.title('Average Sepal Length by Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
sns.histplot(df['sepal_length'], kde=True)
plt.title('Distribution of Sepal Length')
plt.show()
```

```python
sns.boxplot(x='species', y='petal_length', data=df)
plt.title('Petal Length Distribution by Species')
plt.show()
```
