```python
import numpy as np
import pandas as pd
from scipy.stats import ks_2samp
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate synthetic training data (historical)
np.random.seed(42)
n_samples = 1000
train_data = pd.DataFrame({
    'feature_1': np.random.normal(loc=50, scale=10, size=n_samples),
    'feature_2': np.random.uniform(0, 100, size=n_samples),
    'feature_3': np.random.normal(loc=30, scale=5, size=n_samples)
})

# Step 2: Generate synthetic new data (with drift)
new_data = pd.DataFrame({
    'feature_1': np.random.normal(loc=55, scale=10, size=n_samples),  # Mean shifted
    'feature_2': np.random.uniform(10, 90, size=n_samples),           # Range narrowed
    'feature_3': np.random.normal(loc=30, scale=5, size=n_samples)    # No change
})

# Step 3: Compare distributions using Kolmogorov-Smirnov test
print("Drift Detection Report:\n")
for col in train_data.columns:
    stat, p_value = ks_2samp(train_data[col], new_data[col])
    drift = "Detected" if p_value < 0.05 else "Not Detected"
    print(f"{col}: KS Statistic = {stat:.4f}, p-value = {p_value:.4f} --> Drift {drift}")

# Step 4: Visualize distributions
def plot_feature_drift(feature):
    plt.figure(figsize=(10, 4))
    sns.kdeplot(train_data[feature], label='Train', shade=True)
    sns.kdeplot(new_data[feature], label='New', shade=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot all features
for feature in train_data.columns:
    plot_feature_drift(feature)
```
