```python
import pandas as pd
import numpy as np

# ----------------------------------------
# PART 1: Handle Missing Values & Duplicates
# ----------------------------------------

print("=== PART 1: Handling Missing Values & Duplicates ===")

# Step 1: Create Sample Dataset with missing values and duplicates
data1 = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'Eve', None, 'Frank'],
    'Age': [25, np.nan, 30, 25, 45, 35, np.nan],
    'City': ['NY', 'LA', 'NY', 'NY', 'Chicago', 'LA', 'Houston']
})

print("\nOriginal Data:")
print(data1)

# Step 2.1: Identify Missing Values
print("\nMissing Values:\n", data1.isnull().sum())

# Step 2.2: Fill Missing Values
# Fill 'Age' with mean age
data1['Age'] = data1['Age'].fillna(data1['Age'].mean())

# Fill 'Name' with placeholder
data1['Name'] = data1['Name'].fillna('Unknown')

# Step 3.1: Identify Duplicates
print("\nDuplicate Rows:")
print(data1[data1.duplicated()])

# Step 3.2: Remove Duplicates
data1 = data1.drop_duplicates()

print("\nCleaned Data (No Missing Values or Duplicates):")
print(data1)

# ----------------------------------------
# PART 2: Standardization & Formatting Rules
# ----------------------------------------

print("\n=== PART 2: Standardization & Formatting ===")

# Step 1: Create New Sample Dataset
data2 = pd.DataFrame({
    'Product': ['Laptop', 'PHONE', 'Tablet', 'laptop', 'Phone', 'tablet'],
    'Price': [999.999, 499.5, 250.755, 999.999, 499.5, 250.755],
    'Age': [2.9, 1.2, 3.7, 2.9, 1.2, 3.7]
})

print("\nOriginal Product Data:")
print(data2)

# Step 2.1: Standardize Product Names (lowercase)
data2['Product'] = data2['Product'].str.lower()

# Step 2.2: Round 'Age' to nearest integer
data2['Age'] = data2['Age'].round().astype(int)

# Step 2.3: Format 'Price' to 2 decimal places
data2['Price'] = data2['Price'].round(2)

print("\nStandardized & Formatted Product Data:")
print(data2)
```
