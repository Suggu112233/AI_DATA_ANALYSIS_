### Data Drift: Detection Using Statistical Tests
**Question**: Simulate two datasets with a shift in distribution, and apply the Kolmogorov-
Smirnov test to determine if data drift has occurred.

```python
# write your code from here
import numpy as np
from scipy.stats import ks_2samp
import matplotlib.pyplot as plt

# Simulate original dataset (e.g., normal distribution)
np.random.seed(42)
data_original = np.random.normal(loc=0, scale=1, size=1000)

# Simulate new dataset with shifted distribution (mean shifted)
data_shifted = np.random.normal(loc=0.5, scale=1, size=1000)

# Plot histograms to visualize distribution shift
plt.hist(data_original, bins=30, alpha=0.5, label='Original')
plt.hist(data_shifted, bins=30, alpha=0.5, label='Shifted')
plt.legend()
plt.title('Simulated Data Distributions')
plt.show()

# Perform Kolmogorov-Smirnov test
ks_stat, p_value = ks_2samp(data_original, data_shifted)

print(f"KS statistic: {ks_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Data drift detected: distributions differ significantly.")
else:
    print("No significant data drift detected.")

```
