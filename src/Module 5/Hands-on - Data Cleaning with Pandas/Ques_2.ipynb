```python
import pandas as pd

# -----------------------------------------------
# Task 1: Remove Duplicates
# -----------------------------------------------

# Step 1.1: Create synthetic dataset with duplicates
data = pd.DataFrame({
    'ID': [1, 2, 3, 2, 4, 5, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'Bob', 'Eve', 'Frank', 'Frank'],
    'Age': ['25', '30', '35', '30', '40', '45', '45'],  # Age as strings
    'Date_Joined': ['2022-01-15', '2021-07-22', '2020-03-10', '2021-07-22', '2023-06-01', '2022-11-11', '2022-11-11']
})

print("=== Original Data ===")
print(data)

# Step 1.2: Remove duplicate rows
data = data.drop_duplicates()

print("\n=== Data After Removing Duplicates ===")
print(data)

# -----------------------------------------------
# Task 2: Fix Incorrect Data Types
# -----------------------------------------------

# Step 2.1: Convert 'Age' from string to integer
data['Age'] = data['Age'].astype(int)

print("\n=== Data After Fixing Age Data Type (str → int) ===")
print(data.dtypes)

# -----------------------------------------------
# Task 3: Convert String to DateTime
# -----------------------------------------------

# Step 3.1: Convert 'Date_Joined' to datetime
data['Date_Joined'] = pd.to_datetime(data['Date_Joined'])

print("\n=== Data After Converting 'Date_Joined' to datetime ===")
print(data.dtypes)

print("\n=== Final Cleaned Data ===")
print(data)
```
