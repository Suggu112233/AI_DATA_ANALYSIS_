```python
import pandas as pd
import numpy as np

# ------------------------------------------
# Task 1: Detect Missing Data
# ------------------------------------------

# Step 1.1: Create synthetic dataset
data = pd.DataFrame({
    'Name': ['Alice', 'Bob', None, 'Diana', 'Eve', 'Frank', None],
    'Age': [25, np.nan, 30, np.nan, 45, 35, 28],
    'City': ['NY', 'LA', 'Chicago', 'NY', None, 'LA', 'Houston']
})

print("=== Original Data ===")
print(data)

# Step 1.2: Detect missing values
print("\n=== Missing Value Count per Column ===")
print(data.isnull().sum())

print("\n=== Rows with Any Missing Values ===")
print(data[data.isnull().any(axis=1)])

# ------------------------------------------
# Task 2: Handle Missing Data by Dropping
# ------------------------------------------

# Step 2.1: Drop rows with any missing values
dropped_data = data.dropna()

print("\n=== Data After Dropping Rows with Missing Values ===")
print(dropped_data)

# ------------------------------------------
# Task 3: Handle Missing Data by Imputation
# ------------------------------------------

# Step 3.1: Fill missing values
imputed_data = data.copy()

# Fill missing names with 'Unknown'
imputed_data['Name'] = imputed_data['Name'].fillna('Unknown')

# Fill missing age with mean age
imputed_data['Age'] = imputed_data['Age'].fillna(imputed_data['Age'].mean())

# Fill missing city with 'Unknown City'
imputed_data['City'] = imputed_data['City'].fillna('Unknown City')

print("\n=== Data After Imputation ===")
print(imputed_data)
```
