## Implementing ML Model Monitoring Pipelines

### Model Performance Drift:
**Description**: Setup a monitoring pipeline to track key performance metrics (e.g., accuracy, precision) of an ML model over time using a monitoring tool or dashboard.

```python
# write your code from here
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score

# Simulated daily batch predictions and true labels (replace with real data)
daily_batches = [
    {'date': '2025-05-10', 'y_true': [0,1,0,1], 'y_pred': [0,1,0,0]},
    {'date': '2025-05-11', 'y_true': [1,0,1,0], 'y_pred': [1,0,0,0]},
    {'date': '2025-05-12', 'y_true': [0,1,0,1], 'y_pred': [0,1,1,1]},  # degraded precision
]

metrics_list = []

for batch in daily_batches:
    date = batch['date']
    y_true = batch['y_true']
    y_pred = batch['y_pred']

    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, zero_division=0)

    metrics_list.append({'date': date, 'accuracy': accuracy, 'precision': precision})

# Create DataFrame with metrics over time
metrics_df = pd.DataFrame(metrics_list)

print(metrics_df)
import matplotlib.pyplot as plt

metrics_df['date'] = pd.to_datetime(metrics_df['date'])
metrics_df.set_index('date', inplace=True)

plt.figure(figsize=(10,5))
plt.plot(metrics_df.index, metrics_df['accuracy'], marker='o', label='Accuracy')
plt.plot(metrics_df.index, metrics_df['precision'], marker='o', label='Precision')
plt.title('Model Performance Over Time')
plt.xlabel('Date')
plt.ylabel('Metric Value')
plt.ylim(0, 1.1)
plt.legend()
plt.grid(True)
plt.show()

```
