```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset
iris = sns.load_dataset('iris')

# Visualize outliers using boxplot for the 'sepal_length' feature
plt.figure(figsize=(8, 5))
sns.boxplot(x=iris['sepal_length'])
plt.title('Boxplot of Sepal Length - Outlier Detection')
plt.show()
```
