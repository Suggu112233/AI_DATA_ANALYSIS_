## Validate Data Completeness for Critical Fields

**Description**: In `transaction_data.csv` , calculate the missing percentage of critical fields: "transaction_id", "amount", and "date", and visualize them.

```python
# Write your code from here
import pandas as pd
import matplotlib.pyplot as plt

# Simulated transaction_data
data = {
    'transaction_id': [1001, 1002, None, 1004, 1005, None, 1007],
    'amount': [250.0, None, 150.5, 300.0, None, 500.0, 450.0],
    'date': ['2023-01-01', '2023-01-02', '2023-01-03', None, '2023-01-05', '2023-01-06', None],
    'customer_id': [1, 2, 3, 4, 5, 6, 7]  # non-critical field
}

df = pd.DataFrame(data)

# Calculate missing percentages for critical fields
critical_fields = ['transaction_id', 'amount', 'date']
missing_percent = df[critical_fields].isna().mean() * 100

print("Missing Data Percentage for Critical Fields:")
print(missing_percent)

# Visualization
missing_percent.plot(kind='bar', color='skyblue')
plt.title('Missing Data Percentage for Critical Fields')
plt.ylabel('Percentage (%)')
plt.xlabel('Fields')
plt.ylim(0, 100)
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```
